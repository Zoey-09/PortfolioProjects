```python
import yfinance as yf
```

```python
import pandas as pd
sp500 = pd.read_csv('sp500.csv')
sp500
```

```python
sp500 = yf.Ticker('^GSPC')
```

```python
sp500 = sp500.history(period='max')
```

```python
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y='Close', use_index=True)
```

```python
del sp500['Dividends']
del sp500['Stock Splits']
```

```python
#create a new column:Tomorrow
#"Close" column: Tomorrow's price
#shift: pandas method to assist prediction

sp500['Tomorrow'] = sp500["Close"].shift(-1)
sp500
```

```python
#new column "Target"
#astype: converting outcome into integer
sp500['Target'] = (sp500['Tomorrow'] > sp500['Close']).astype(int)
sp500
```

```python
#remove data before 1990-01-01
#.copy(): return a new list that does not modify the original list
sp500 = sp500.loc['1990-01-01':].copy()
sp500
```

```python
#Training initial Machine Learning Model
#n_estimators: the no. of individual desision tree we want to train.
#min_samples_split: the min. no. of sample required for a split.



from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = sp500.iloc[:-100] #iloc[:-100]: split data and take all of it except last 100 rows
test = sp500.iloc[-100:]  #iloc[-100:]: split data only take last 100 rows

predictors = ["Close", "Volume", "Open", "High", "Low"] #create columns for 'predictors'.
model.fit(train[predictors], train["Target"]) #.fit():After model is initialized, sklearn fit method uses the training data as an input to train the machine learning model.
```

```python
#measuring model accuracy.
from sklearn.metrics import precision_score

preds = model.predict(test[predictors]) #generating prediction
preds = pd.Series(preds, index=test.index) #automatically generating index for the data set.
precision_score(test["Target"], preds) # comparing actual Target against preds target.
```

```python
preds = pd.Series(preds, index=test.index)
preds
```

```python
#pd.concat: pandas func to merge files. axis = 0 (file 1 ↓ file 2 ), axis = 1 (file 1 → file 2)
#.plot(): pandas method to create diagrams, plot a series of dataframe
combined = pd.concat([test["Target"], preds], axis=1) #.concat: merge files side by side(axis=1)
combined.plot()
```

```python
#building a Backtesting System
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)#append all prediction
        
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.6] = 1
    preds[preds <.6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions['Predictions'])
```
